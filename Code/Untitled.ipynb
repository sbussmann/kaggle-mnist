```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score
```

```python
dloc = '../Data/'
```

```python
train = pd.read_csv(dloc+'train.csv')
```

```python
Xcol = train.columns[1:]
ycol = 'label'
X = train[Xcol]
y = train[ycol]
```

```python
clf = RandomForestClassifier()
```

```python
scores = cross_val_score(clf, X, y, cv=5)
print(scores, scores.mean(), scores.std())
```

```python
clf.fit(X, y)
```

```python
test = pd.read_csv(dloc + 'test.csv')
```

```python
Xtest = test
```

```python
ypredict = clf.predict(Xtest)
```

```python
ypredict
```

```python
dfpredict = pd.DataFrame(ypredict)
```

```python
dfpredict.columns = ['Label']
```

```python
dfpredict.to_csv(dloc + 'predict_RFbenchmark.csv', index_label='ImageId')
```
